```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
# Step 1: Click the button
button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CLASS_NAME, "wofuke"))
)
button.click()

# Step 2: Wait for the content to appear
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, '[data-hook="info-section-description"]'))
)

# Step 3: Parse and extract text
soup = BeautifulSoup(driver.page_source, "html.parser")
info = soup.select_one('[data-hook="info-section-description"]')

if info:
    print(info.text.strip())
else:
    print("Content not found.")

driver.quit()
```

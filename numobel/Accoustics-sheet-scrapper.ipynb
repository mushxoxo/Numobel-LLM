```python
import requests
import subprocess
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
url = "https://www.numobel.in/shop-online"

webpage = subprocess.run(['curl', '--ipv4', '-v', url], capture_output=True, text=True).stdout
soup = BeautifulSoup(webpage, 'lxml')
```

```python
products = soup.find_all("div", class_ = "ejYUwA t1Z_7C")
```

```python
for product in products:
    product_link = product.find("a", class_ = "AJctir bGFTjD qc8uk6")
    product_link = product_link["href"]
    print(product_link)
```

```python
product_link = "https://www.numobel.in/product-page/numobelacoustics-pet-acoustic-sheets-skins"
```

```python
driver = webdriver.Chrome()   # make sure chromedriver is installed
driver.get(product_link)

time.sleep(5)  # wait for JS to load
```

```python
name = driver.find_element(By.CSS_SELECTOR, '[data-hook="product-title"]').text
print(name)
```

```python
price = driver.find_element(By.CSS_SELECTOR, '[data-hook="product-price"]').text
```

```python
price
```

```python
formatted_price = driver.find_element(By.CSS_SELECTOR, '[data-hook="formatted-primary-price"]').text
```

```python
formatted_price
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)

dropdown = wait.until(
    EC.element_to_be_clickable(
        (By.CSS_SELECTOR, '[data-hook="core-dropdown"]')
    )
)
```

```python
dropdown.click()
```

```python
options = wait.until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, '[data-hook="dropdown-option"]')
    )
)

sizes = [opt.text for opt in options]
print("Available Sizes:", sizes)
```

```python
dropdown.click()
```

```python
color_elements = wait.until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, 'input[data-hook="hidden-radio"]')
    )
)

colors = [c.get_attribute("aria-label") for c in color_elements]

print("Available Colors:", colors)
```

```python
thumbnail_imgs = wait.until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, 'button[data-hook="thumbnail-root"] img')
    )
)

image_urls = [img.get_attribute("src") for img in thumbnail_imgs]

print("Thumbnail Images:")
for url in image_urls:
    print(url)
```

```python
description_element = wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, '[data-hook="description"]')
    )
)

description_text = description_element.text

print(description_text)
```

```python
driver.quit()
```
